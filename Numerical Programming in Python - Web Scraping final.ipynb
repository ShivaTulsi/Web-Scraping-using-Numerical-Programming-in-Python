## **Scrapping Movies Data**

````python
# Specifying the URL from which movies related data will be fetched
url='https://www.justwatch.com/in/movies?release_year_from=2000'

# Define headers to mimic a web browser
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

# Sending an HTTP GET request to the URL with headers
page=requests.get(url, headers=headers)
# Parsing the HTML content using BeautifulSoup with the 'html.parser'
soup=BeautifulSoup(page.text,'html.parser')
# Printing the prettified HTML content
print(soup.prettify())
````

## **Fetching Movie URL's**

````python
# Write Your Code here
movie_url=[]
# Find the elements that contain the movie URLs
movie_container=soup.find_all('a',class_='title-list-grid__item--link')
# Extract and print the URLs
for movie in movie_container:
       movie_url.append(movie['href'])
print( movie_url)
````

````python
#importing library for joining url
from urllib.parse import urljoin
base_url='https://www.justwatch.com/in/'

movies=[]
for relative_url in movie_url :
    absolute_url = urljoin(base_url, relative_url)
    movies.append(absolute_url)

# Print the list of absolute movie URLs
print(movies)
````

## **Scrapping Movie Title**

````python
# Write Your Code here
movie_title=[]
# Extract and print the movie names
for movie in movies:
  movie_title.append(movie.split('/')[-1])

# Print the extracted movie names
print(movie_title)
````

## **Scrapping release Year**

````python
# Write Your Code here
release_years=[]
# Define headers to mimic a web browser
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

for movie_url in movies:
  # Fetch the HTML content of the movie URL using requests with headers
    response=requests.get(movie_url, headers=headers)
    if response.status_code==200:
        # Parse the HTML content with BeautifulSoup
        soup=BeautifulSoup(response.content, 'html.parser')
        try:
            # Find the release year element
            release_year_element=soup.find('span', class_='text-muted')
            release_year=release_year_element.get_text(strip=True).strip("()")
        except AttributeError:
            release_year='N/A'
        release_years.append({'Movie URL': movie_url, 'Release Year': release_year})
    else:
        print(f"Failed to fetch the HTML content for {movie_url}")

# Print the final result
for movie_data in release_years:
    print(movie_data)
````

## **Scrapping Genres**

````python
# Write Your Code here
# Movie genre

movie_genre_list = []# List to store all movie genre

# For every movie title present in movies_title_list , Finding their genre
for movie in movie_title_list:

    # Constructing the absolute URL for fetching each movie
    absolute_url = 'https://www.justwatch.com/in/movie/' + movie
    response_g = requests.get(absolute_url)
    soup = BeautifulSoup(response_g.text,'html.parser')

    # Selecting only those h3 whose heading is genres
    h3_element = soup.find('h3', class_='detail-infos__subheading', string='Genres')

    if h3_element:
        # Check if the next sibling is a div with class "detail-infos__value"
        div_element = h3_element.find_next_sibling('div', class_='detail-infos__value')

        if div_element:
            movie_genre_list.append(div_element.text.strip())
        else:
            movie_genre_list.append("Genre Not Listed")
    else:
         movie_genre_list.append("Genre Not Listed")

````

## **Scrapping IMBD Rating**

````python
# Write Your Code here
# Movie Imdb Rating

movie_imdb_list = []# List to store all movie imdb rating

# For every movie title present in movies_title_list , Finding their Imdb Rating
for movie in movie_title_list:

    # Constructing the absolute URL for fetching each movie
    absolute_url = 'https://www.justwatch.com/in/movie/' + movie
    response_g = requests.get(absolute_url)
    soup = BeautifulSoup(response_g.text, 'html.parser')

    # Selecting only those h3 whose heading is Rating
    h3_element = soup.find('h3', class_='detail-infos__subheading', string='Rating')

    if h3_element:
        # Check if the next sibling is a div with class "detail-infos__value"
        div_element = h3_element.find_next_sibling('div', class_='detail-infos__value')

        if div_element:
            inside_div = div_element.find_all('div', class_='jw-scoring-listing__rating')

            # Check if inside_div is non-empty
            if inside_div:
                inside_div_last = inside_div[-1] # extracting last div of inside div as last div elemnt contains span (in which last span contains rating)

                # Check if inside_div_last is non-empty
                if inside_div_last:
                    span_all = inside_div_last.find_all('span')

                    # Check if span_all is non-empty
                    if span_all:
                        span_last = span_all[-1] # Here we are extracting rating from the last span(span_last) inside last div(inside_div_last) of main div_element(div_element)
                        movie_imdb_list.append(span_last.text.strip())
                    else:
                        movie_imdb_list.append("Imdb Rating Not Listed.")
                else:
                    movie_imdb_list.append("Imdb Rating Not Listed.")
            else:
                movie_imdb_list.append("Imdb Rating Not Listed.")
        else:
            movie_imdb_list.append("Imdb Rating Not Listed.")
    else:
        movie_imdb_list.append("Imdb Rating Not Listed.")

````

````python
# Filtering movies and TV shows to include only those released in the last two years and with an IMDB Rating of 7 or higher.

from datetime import datetime, timedelta

# Get the current date
current_date = datetime.now()

# Calculate the date 2 years ago from the current date
two_years_ago = current_date - timedelta(days=365 * 2)

def filter_df(df, release_year_col, imdb_rating_col, years_ago, current_date):
    # Convert 'Release Year' to datetime format
    df[release_year_col] = pd.to_datetime(df[release_year_col], errors='coerce')

    # Filter the DataFrame to include only entries released in the last `years_ago` years
    filtered_df = df[(df[release_year_col] >= two_years_ago) & (df[release_year_col] <= current_date)].copy()

    # Converting 'IMDB Rating' column to a string so that, in the next step, we can convert it to numeric values
    filtered_df.loc[:, imdb_rating_col] = filtered_df[imdb_rating_col].astype(str)

    # Extract numeric part and convert to numeric
    filtered_df[imdb_rating_col] = pd.to_numeric(filtered_df[imdb_rating_col].str.extract(r'([\d.]+)', expand=False), errors='coerce')

    # Filter the DataFrame to include only entries whose IMDb Rating >= 7
    filtered_df = filtered_df[filtered_df[imdb_rating_col] >= 7]

    return filtered_df

# Filtering Movies
filtered_df_movies = filter_df(df_movies, 'Release Year', 'IMDB Rating', two_years_ago, current_date)

# Filtering TV Shows
filtered_df_tv_shows = filter_df(df_tv_shows, 'Release Year', 'IMDB Rating', two_years_ago, current_date)

````

````python
# Filtering movies and TV shows to include only those released in the last two years and with an IMDB Rating of 7 or higher.

from datetime import datetime, timedelta

# Get the current date
current_date = datetime.now()

# Calculate the date 2 years ago from the current date
two_years_ago = current_date - timedelta(days=365 * 2)

def filter_df(df, release_year_col, imdb_rating_col, years_ago, current_date):
    # Convert 'Release Year' to datetime format
    df[release_year_col] = pd.to_datetime(df[release_year_col], errors='coerce')

    # Filter the DataFrame to include only entries released in the last `years_ago` years
    filtered_df = df[(df[release_year_col] >= two_years_ago) & (df[release_year_col] <= current_date)].copy()

    # Converting 'IMDB Rating' column to a string so that, in the next step, we can convert it to numeric values
    filtered_df.loc[:, imdb_rating_col] = filtered_df[imdb_rating_col].astype(str)

    # Extract numeric part and convert to numeric
    filtered_df[imdb_rating_col] = pd.to_numeric(filtered_df[imdb_rating_col].str.extract(r'([\d.]+)', expand=False), errors='coerce')

    # Filter the DataFrame to include only entries whose IMDb Rating >= 7
    filtered_df = filtered_df[filtered_df[imdb_rating_col] >= 7]

    return filtered_df

# Filtering Movies
filtered_df_movies = filter_df(df_movies, 'Release Year', 'IMDB Rating', two_years_ago, current_date)

# Filtering TV Shows
filtered_df_tv_shows = filter_df(df_tv_shows, 'Release Year', 'IMDB Rating', two_years_ago, current_date)
````

````python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import numpy as np
import re

# Selenium imports
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
import time

# Function to set up the headless Chrome driver
def setup_driver():
    chrome_options = Options()
    chrome_options.add_argument("--headless")
    chrome_options.add_argument("--no-sandbox")
    chrome_options.add_argument("--disable-dev-shm-usage")
    chrome_options.add_argument("--disable-gpu")
    chrome_options.add_argument("--window-size=1920,1080")
    chrome_options.add_argument("--disable-extensions")
    chrome_options.add_argument("--proxy-server='direct://'")
    chrome_options.add_argument("--proxy-bypass-list='*' ")
    chrome_options.add_argument("--start-maximized")

    # ChromeDriverManager will automatically detect the installed Google Chrome version
    # and download the compatible ChromeDriver.
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service, options=chrome_options)
    return driver
````

````python
# URL from which movies data is fetched
url="https://www.justwatch.com/in/movies?release_year_from=2000"

driver = setup_driver()
driver.get(url)

# Wait for the initial page content to load
time.sleep(3) # A brief pause to allow initial content to load

# List to store unique movie titles
movie_title_list = []

# Loop to load more content until at least 50 titles are collected
while len(movie_title_list) < 50:
    # Get page source after initial load or scroll
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    movie_titles_on_page = soup.find_all('a', class_='title-list-grid__item--link', attrs={'href': True})
    initial_count = len(movie_title_list)

    for movie_title_element in movie_titles_on_page:
        data_id_value = movie_title_element['href']
        # Extract title and add if not already present
        if data_id_value.startswith("/in/movie/"):
            title_part = data_id_value.replace("/in/movie/", "").upper()
            if title_part not in movie_title_list:
                movie_title_list.append(title_part)

    # Check if we have collected enough titles
    if len(movie_title_list) >= 50:
        break

    # If no new titles were added on this iteration, and we haven't reached 50,
    # it might mean there's no more content or the "Load More" button isn't working.
    if len(movie_title_list) == initial_count:
        print(f"No new movie titles found after attempting to scroll. Current count: {len(movie_title_list)}")
        break # Exit if no progress

    # Try to find and click the "Load More" button
    try:
        load_more_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//button[contains(., 'Load More')]"))
        )
        driver.execute_script("arguments[0].click();", load_more_button)
        time.sleep(3) # Wait for new content to load
    except:
        print("No 'Load More' button found or clickable for movies, or all content loaded.")
        break # Exit if button not found or not clickable

print(f"Scraped {len(movie_title_list)} movie titles (first 50 or all available if less than 50).")
driver.quit() # Close the browser
````

````python
# Filtering movies and TV shows to include only those released in the last two years and with an IMDB Rating of 7 or higher.

from datetime import datetime, timedelta

# Get the current date
current_date = datetime.now()

# Calculate the date 2 years ago from the current date
two_years_ago = current_date - timedelta(days=365 * 2)

def filter_df(df, release_year_col, imdb_rating_col, years_ago, current_date):
    # Convert 'Release Year' to datetime format
    df[release_year_col] = pd.to_datetime(df[release_year_col], errors='coerce')

    # Filter the DataFrame to include only entries released in the last `years_ago` years
    filtered_df = df[(df[release_year_col] >= two_years_ago) & (df[release_year_col] <= current_date)].copy()

    # Converting 'IMDB Rating' column to a string so that, in the next step, we can convert it to numeric values
    filtered_df.loc[:, imdb_rating_col] = filtered_df[imdb_rating_col].astype(str)

    # Extract numeric part and convert to numeric
    filtered_df[imdb_rating_col] = pd.to_numeric(filtered_df[imdb_rating_col].str.extract(r'([\d.]+)', expand=False), errors='coerce')

    # Filter the DataFrame to include only entries whose IMDb Rating >= 7
    filtered_df = filtered_df[filtered_df[imdb_rating_col] >= 7]

    return filtered_df

# Filtering Movies
filtered_df_movies = filter_df(df_movies, 'Release Year', 'IMDB Rating', two_years_ago, current_date)

# Filtering TV Shows
filtered_df_tv_shows = filter_df(df_tv_shows, 'Release Year', 'IMDB Rating', two_years_ago, current_date)
````

````python
# URL from which tv shows data is fetched
tv_url='https://www.justwatch.com/in/tv-shows?release_year_from=2000'

driver_tv = setup_driver()
driver_tv.get(tv_url)

# Wait for the initial page content to load
time.sleep(3) # A brief pause to allow initial content to load

# List to store unique TV show titles
tv_show_title_list = []

# Loop to load more content until at least 50 titles are collected
while len(tv_show_title_list) < 50:
    # Get page source after initial load or scroll
    soup_tv = BeautifulSoup(driver_tv.page_source, 'html.parser')

    tv_show_titles_on_page = soup_tv.find_all('a', class_='title-list-grid__item--link', attrs={'href': True})
    initial_count = len(tv_show_title_list)

    for tv_show_title_element in tv_show_titles_on_page:
        data_id_value = tv_show_title_element['href']
        # Extract title and add if not already present
        if data_id_value.startswith("/in/tv-show/"):
            title_part = data_id_value.replace("/in/tv-show/", "").upper()
            if title_part not in tv_show_title_list:
                tv_show_title_list.append(title_part)

    # Check if we have collected enough titles
    if len(tv_show_title_list) >= 50:
        break

    # If no new titles were added on this iteration, and we haven't reached 50,
    # it might mean there's no more content or the "Load More" button isn't working.
    if len(tv_show_title_list) == initial_count:
        print(f"No new TV show titles found after attempting to scroll. Current count: {len(tv_show_title_list)}")
        break # Exit if no progress

    # Try to find and click the "Load More" button
    try:
        load_more_button_tv = WebDriverWait(driver_tv, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//button[contains(., 'Load More')]"))
        )
        driver_tv.execute_script("arguments[0].click();", load_more_button_tv)
        time.sleep(3) # Wait for new content to load
    except:
        print("No 'Load More' button found or clickable for TV shows, or all content loaded.")
        break # Exit if button not found or not clickable

print(f"Scraped {len(tv_show_title_list)} TV show titles (first 50 or all available if less than 50). Jard'in).")
driver_tv.quit() # Close the browser
````
